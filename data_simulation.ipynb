```python
import pandas as pd
import numpy as np
import names
import random
```

```python
n = 40

names_generated = []
values_generated = []
groups_generated = []

groups_values = ["PLACEBO", "EXPERIMENTAL"]
mu_exp, sigma_exp = 12, 3
mu_placebo, sigma_placebo = 20, 5

for i in range(n):
    name = names.get_full_name()
    group = random.choice(groups_values)
    if group == groups_values[1]:
        value = int(np.random.normal(mu_exp, sigma_exp, 1)[0])
    else:
        value = int(np.random.normal(mu_placebo, sigma_placebo, 1)[0])
    names_generated.append(name)
    groups_generated.append(group)
    values_generated.append(value)


print(groups_generated, values_generated, names_generated)
```

```python

arr = [[names_generated[i], values_generated[i], groups_generated[i]] for i in range(n)]
df = pd.DataFrame(data=arr, columns=["nombre", "dias_hospitalizacion", "grupo"])
df
```

```python
df.to_csv("estudio.csv", sep=";")
```
